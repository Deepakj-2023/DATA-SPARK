```python
def convert_date_process(date_str):
    try:
        return pd.to_datetime(date_str, format='%m/%d/%Y', errors='coerce').strftime('%m/%d/%Y')
    except ValueError:
        try:
            return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce').strftime('%m/%d/%Y')
        except ValueError:
            return date_str 
customers['Birthday'] = customers['Birthday'].apply(convert_date_process)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection details
host = "localhost"
port = "3306"
username = "root"
password = "1914Deepak"
database = "capstone_data"

# SQLAlchemy engine
engine_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)

# Define table names and their corresponding DataFrames
dataframes = {
    "overall_data": overall_data,
    "customers": customers,
    "products": products,
    "exchangerates": exchangerates,
    "stores": stores
}

# Loop through each table and export to SQL
for table_name, df in dataframes.items():
    if df is not None and not df.empty:  # Check if DataFrame is valid
        try:
            df.to_sql(table_name, engine, if_exists="replace", index=False)
            print(f"Successfully imported {table_name} to SQL.")
        except Exception as e:
            print(f"Failed to import {table_name} to SQL: {e}")
    else:
        print(f"DataFrame for {table_name} is empty or undefined.")
```
